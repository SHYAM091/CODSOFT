```python
import pandas as pd
import numpy as np

# Load the dataset with an appropriate encoding
try:
    df = pd.read_csv(r'S:\dataset intern\spam1 train.csv', encoding='ISO-8859-1')
except UnicodeDecodeError:
    df = pd.read_csv(r'S:\dataset intern\spam1 train.csv', encoding='cp1252')

# Display first few rows to understand the dataset
print(df.head())

# 1. Remove or fill null values
# Drop rows with any null values
df_cleaned = df.dropna()

# Alternatively, fill null values with a placeholder, e.g., an empty string for text columns
# df_filled = df.fillna('')

# 2. Preprocessing
# Example: if the dataset has a text column named 'text', preprocess it
# Convert text to lowercase
if 'text' in df_cleaned.columns:
    df_cleaned['text'] = df_cleaned['text'].str.lower()

    # Remove punctuation
    df_cleaned['text'] = df_cleaned['text'].str.replace('[^\w\s]', '', regex=True)

# Example: if you have numerical columns, you might want to normalize or standardize them
# Assuming 'feature1' and 'feature2' are numerical columns
numerical_columns = ['feature1', 'feature2']
for col in numerical_columns:
    if col in df_cleaned.columns:
        # Normalize
        df_cleaned[col] = (df_cleaned[col] - df_cleaned[col].mean()) / df_cleaned[col].std()

# Display cleaned and preprocessed dataset
print(df_cleaned.head())
```
